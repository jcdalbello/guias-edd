### Ejercicio 6

Sea $G$ un grafo no dirigido, utilizar el recorrido DFS para determinar si $G$ es bipartito.

```python
from edd.cola import Cola
from edd.grafo import Grafo, Vertice

def es_bipartito_con_dfs(G: Grafo):  # agregar argumentos necesarios.
    s = list(G.vertices)[0]
    color = {}
    visitado = {v:False for v in G.vertices}
    
    color[s] = True

    def DFS(v: Vertice, visitado, color):
        visitado[v] = True
        
        for w in v.adyacentes:
            if not visitado[w]:
                color[w] = not color[v]
                DFS(w, visitado, color)
            else:
                if color[w] == color[v]:
                    return False
        return True

    return DFS(s, visitado, color)

Grafo_no_bipartito = Grafo()
Grafo_no_bipartito.agregar_arista("E", "D")
Grafo_no_bipartito.agregar_arista("D", "A")
Grafo_no_bipartito.agregar_arista("E", "F")
Grafo_no_bipartito.agregar_arista("E", "H")
Grafo_no_bipartito.agregar_arista("E", "G")
Grafo_no_bipartito.agregar_arista("D", "G")
Grafo_no_bipartito.agregar_arista("D", "B")
Grafo_no_bipartito.agregar_arista("A", "B")
Grafo_no_bipartito.agregar_arista("A", "C")
Grafo_no_bipartito.agregar_arista("I", "C")

Grafo_bipartito = Grafo()
Grafo_bipartito.agregar_arista("E", "F")
Grafo_bipartito.agregar_arista("E", "H")
Grafo_bipartito.agregar_arista("E", "G")
Grafo_bipartito.agregar_arista("D", "G")
Grafo_bipartito.agregar_arista("D", "B")
Grafo_bipartito.agregar_arista("A", "B")
Grafo_bipartito.agregar_arista("A", "C")
Grafo_bipartito.agregar_arista("I", "C")

print(f"Grafo_no_bipartito -> {es_bipartito_con_dfs(Grafo_no_bipartito)}")
print(f"Grafo_bipartito -> {es_bipartito_con_dfs(Grafo_bipartito)}")
```

### Ejercicio 7

Sea $G$ un grafo no dirigido, escribir un algoritmo que permita detectar si tiene un ciclo.

1. ¬øQu√© tipo de recorrido se utiliz√≥?
2. ¬øEs posible resolverlo con otro recorrido?

```python
from edd.grafo import Grafo, Vertice

def contiene_ciclos(G: Grafo):  # agregar argumentos necesarios.
    s = list(G.vertices)[0]
    visitado = {v:False for v in G.vertices}
    
    def DFS(v: Vertice, visitado, previo=None):
        visitado[v] = True
        for w in v.adyacentes:
            if not visitado[w]:
                DFS(w, visitado, v)
            else:
                if v != previo:
                    return False
        return True

    return DFS(s, visitado)


grafo_sin_ciclos = Grafo()
grafo_sin_ciclos.agregar_arista("A", "B")
grafo_sin_ciclos.agregar_arista("B", "C")
grafo_sin_ciclos.agregar_arista("B", "D")

grafo_con_ciclos = Grafo()
grafo_con_ciclos.agregar_arista("A", "B")
grafo_con_ciclos.agregar_arista("B", "C")
grafo_con_ciclos.agregar_arista("B", "D")
grafo_con_ciclos.agregar_arista("A", "D")
```

1. Para resolver el problema, se uso un algoritmo basado en el algoritmo de recorrido DFS.
2. Tambien podria resolverse con un algoritmo de ordenamiento topologico, si este lanza una excepcion entonces existe al menos un ciclo en el grafo.
