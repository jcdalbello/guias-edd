### Ejercicio 4

Escribir un método que imprima por pantalla el camino mínimo y el costo para llegar a cada vértice luego de ejecutar los algoritmos de camino mínimo.

```python
from edd.grafo import DiGrafo, Vertice
from edd.pila import Pila
from math import inf


def imprimir_caminos(self, s: Vertice) -> None:
    distancia = {}
    previo = {}

    if any(arista.peso < 0 for arista in self.aristas):
        distancia, previo = self.bellman_ford(s)
    else:
        distancia, previo = self.dijkstra(s)

    def _camino_desde(v_inicio: Vertice, v_destino: Vertice, previo: dict[Vertice, Vertice | None]) -> list[Vertice]:
        def _camino_inverso(v_inicio: Vertice, v_destino: Vertice, previo: dict[Vertice, Vertice | None], camino: list[Vertice] = []) -> list[Vertice]:
            camino.append(v_destino)
            if v_destino != v_inicio:
                _camino_inverso(v_inicio, previo[v_destino], previo, camino)
            return camino
                
        return list(reversed(_camino_inverso(v_inicio, v_destino, previo)))

    for v in self.vertices:
        # print(f"{v} : distancia={distancia[v]}, camino={_camino_desde(s, v, previo)}")
        camino = ""
        for w in _camino_desde(s, v, previo):
            camino += w.id + "-"
        camino = camino[:-1]
        
        print(f"{camino} {distancia[v]}")


DiGrafo.imprimir_caminos = imprimir_caminos


# Salida de ejemplo:
# A-D-C 15
# A-B 7
# A-B-E 10
# A-D 3
```

### Ejercicio 10

Modificar el algoritmo de Dikstra para que en el caso de que haya más de un camino mínimo hasta un vértice elija el que atraviesa la menor cantidad de aristas

```python
from edd.coladeprioridad import ColaDePrioridad
from math import inf


def dijkstra_cantidad_aristas(self, s: Vertice | str) -> tuple[dict[Vertice, float], dict[Vertice, Vertice | None]]:
    s = self._vertices[str(s)]

    pq: ColaDePrioridad = ColaDePrioridad()

    distancia: dict[Vertice, float] = {v: inf for v in self.vertices}
    previo: dict[Vertice, Vertice | None] = {v: None for v in self.vertices}
    visitado: dict[Vertice, bool] = {v: False for v in self.vertices}
    cantidad_aristas: dict[Vertice, int] = {v:inf for v in self.vertices}

    distancia[s] = 0
    cantidad_aristas[s] = 0
    pq.encolar(s, distancia[s])

    while not pq.esta_vacia():
        v, _ = pq.desencolar_minimo()

        visitado[v] = True

        for _, w, peso_v_w in v.aristas:
            if not visitado[w] and distancia[v] + peso_v_w < distancia[w]:
                distancia[w] = distancia[v] + peso_v_w
                previo[w] = v
                pq.encolar(w, distancia[w])

            if not visitado[w]:
                if distancia[v] + peso_v_w < distancia[w] or (distancia[v] + peso_v_w < distancia[w] and cantidad_aristas[v] + 1 < cantidad_aristas[w]):
                    distancia[w] = distancia[v] + peso_v_w
                    previo[w] = v
                    cantidad_aristas[w] = cantidad_aristas[v] + 1
                    pq.encolar(w, distancia[w])

    return distancia, previo
```
